```python
def damn_interesting(num1, num2=None, kind="trivia"):
    if num2 is None:
        num2 = ""
    elif (kind != "date"):
        raise ValueError("The value num2 is only valid for dates (kind='date')")
    else:
        num2 = "/" + str(num2)
    kind = "/" + kind
    reszta = str(num1) + num2 + kind
    link = "http://numbersapi.com/" + reszta + "?json"
    www = requests.get(link)
    wynik = json.loads(www.text)
    return wynik["text"]
```

```python
damn_interesting(5, kind="math")
```

```python
damn_interesting(1989, kind="year")
```

```python
damn_interesting(32, kind="date")
```

```python
damn_interesting(3, 2, kind="date")
```

# Dla jakich liczb nie ma ciekawostek?

* można dla każdej pojedynczej liczby robić zapytanie
* a można przyjrzeć się działowi Batch Requests

```python
for typ in ["math", "trivia", "date", "year"]:
    for liczba in range(500, 700):
        www = requests.get(f"http://numbersapi.com/{liczba}/{typ}?json")
        try:
            wynik = json.loads(www.text)
            if not wynik["found"]:
                print(f"Nie znaleziono {liczba}-{typ}")
        except json.JSONDecodeError:
            print(www.text)
        
```

```python
def no_trivia(low, up, kind="trivia"):
    reszta = "/" + kind
    link = "http://numbersapi.com/" + str(low) + ".." + str(up) + reszta + "?json"
    www = requests.get(link)
    plikjson = json.loads(www.text)
    listawynikowa = []
    for i in range(low, up+1):
        if not str(i)['found']:
            listawynikowa.append(i)
    if len(listawynikowa) == 0:
        print("Brak wyników")
    else:
        return listawynikowa
```

# Porządnie, batchami:

```python
start = 0
end = 2000

for typ in ["math", "trivia", "date", "year"]:
    for liczba in range(start, end, 100):
        try:
            if liczba % 25 == 0:
                print(f"Jesteśmy w momencie: {typ} - {liczba}")
            www = requests.get(f"http://numbersapi.com/{liczba}..{liczba+100}/{typ}?json")
            j = json.loads(www.text)
            for liczba2 in range(liczba, liczba+100):
                if not j[str(liczba2)]["found"]:
                    print(f"Nie znaleziono: {typ} - {liczba2}")
        except json.JSONDecodeError:
            print(www.text)
```

```python
from tqdm.notebook import tqdm

start = 0
end = 2000

for typ in ["math", "trivia", "date", "year"]:
    for liczba in tqdm(range(start, end, 100)):
        try:
            if liczba % 25 == 0:
                print(f"Jesteśmy w momencie: {typ} - {liczba}")
            www = requests.get(f"http://numbersapi.com/{liczba}..{liczba+100}/{typ}?json")
            j = json.loads(www.text)
            for liczba2 in range(liczba, liczba+100):
                if not j[str(liczba2)]["found"]:
                    print(f"Nie znaleziono: {typ} - {liczba2}")
        except json.JSONDecodeError:
            print(www.text)
```
