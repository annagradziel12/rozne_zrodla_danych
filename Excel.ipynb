# Wczytujemy arkusz "wg województw"

* pomijamy kolumny z numerami identyfikacyjnymi placówek oświatowych i województw
* chcemy mieć inaczej zgrupowane dane: najpierw duża, zewnętrzna grupa "przedszkole" a potem wyszczególnione województwa

```python
wypadki2 = pd.read_excel("WypadkiwgwojewodztwIX2016.xls",
                         sheet_name="wg województw", 
                         header=2, 
                         index_col=[1,0], 
                         skipfooter=1, 
                         usecols="B,D:Q")
wypadki2.sort_index(axis=0, level=0, inplace=True)
wypadki2
```
