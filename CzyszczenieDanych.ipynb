# Zmieniamy nazwy, usuwamy kolumny

````python
oceny2 = oceny.copy()
oceny2.drop("Z15(5)", axis=1, inplace=True)
oceny2.rename({"Z13(5)": "Z13(4+1)"}, axis=1, inplace=True)
oceny2.head()
````

````python
numer_zadania, pkt_lab, pkt_dom = re.findall("Z(\d+)\((\d+)\+(\d+)\)", oceny2.columns[2])[0]
rozdzielone_punkty = oceny2.iloc[:,2].str.split("+", expand=True)

oceny2[f"L{numer_zadania}({pkt_lab})"] = rozdzielone_punkty[0]
oceny2[f"D{numer_zadania}({pkt_dom})"] = rozdzielone_punkty[1]
oceny2
````

````python
oceny2 = oceny.copy()
oceny2.drop("Z15(5)", axis=1, inplace=True)
oceny2.rename({"Z13(5)": "Z13(4+1)"}, axis=1, inplace=True)
do_usuniecia = []

for indeks, nazwa_kolumny in enumerate(oceny2.columns):
    dopasowanie = re.findall("Z(\d+)\((\d+)\+(\d+)\)", nazwa_kolumny)
    # print(nazwa_kolumny)
    if len(dopasowanie) == 0:
        continue
    numer_zadania, pkt_lab, pkt_dom = dopasowanie[0]
    rozdzielone_punkty = oceny2.iloc[:,indeks].str.split("+", expand=True)
    oceny2[f"L{numer_zadania}({pkt_lab})"] = rozdzielone_punkty[0]
    oceny2[f"D{numer_zadania}({pkt_dom})"] = rozdzielone_punkty[1]
    do_usuniecia.append(nazwa_kolumny)
    
for kolumna in do_usuniecia:
    oceny2.drop(kolumna, axis=1, inplace=True)
    
oceny2
````

# Wystawianie ocen

* policzyć maksymalną liczbę punktów do zdobycia
* zsumować punkty dla każdego studenta
* policzyć ile procent punktów zdobył każdy student
* oceny wystawiamy tak:
* [0, 0.5) -> 2
* [0.5, 0.6) -> 3
* [0.6, 0.7) -> 3.5
* [0.7, 0.8) -> 4
* [0.8, 0.9) -> 4.5
* [0.9, 1.0) -> 5

````python
oceny5 = oceny4.copy()
oceny5["Suma"] = oceny5.sum(axis=1)
oceny5
````

````python
maksy = oceny5.columns.str.extract("((?:L|D)\d+\((\d+)\))") # (L|D) nawiasy zeby ograniczyc alternatywe, ?: zeby powiedziec, ze te nawiasy nie sa capturing group
maks = maksy[1].astype(float).sum()
maks
````

````python
oceny5["Procenty"] = oceny5["Suma"]/maks
oceny5
````
